```python
__author__ = 'Solution by Linda Jaber, CUSP NYU 2018, Principles of Urban Informatics'

# compatible with pyhton 2 and python 3

try:
    import urllib2 as ulib
except ImportError:
    import urllib as ulib

try:
    from StringIO import BytesIO as io
except ImportError:
    from io import BytesIO as io

try:
    from urllib import urlencode as urlencode
except ImportError:
    from urllib.parse import urlencode as urlencode
    
try:
    from urllib import urlopen as urlopen
except ImportError:
    from urllib.request import urlopen as urlopen
    
try:
    from urllib2 import HTTPError as HTTPError
except ImportError:
    from urllib.error import HTTPError as HTTPError
    
import ast
import pandas as pd
```

```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q='

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urlencode({'format': format, 'q': query}).encode('utf-8')
    try:
        response =urlopen(source, data)
    except HTTPError as e:
        raise (ValueError('\n'.join(ast.literal_eval(e.readline())['error'])))
    
    return response.read()

def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))
```

```python
query = '''
SELECT * 
FROM fb55.citibike
LIMIT 5
'''
get_data(query)
```

## Task 1 — Familiarize with SQL Clauses
#### Sort data by start_station_id, tripduration 
- Only checking trips with duration <= 3 hours 

```python
query = ''' 
SELECT * 
FROM fb55.citibike
WHERE tripduration <= 10800
ORDER BY start_station_id, tripduration
LIMIT 5
'''
get_data(query)
```

#### Only show the top/last 10 records (aka head and tail in SQL) 

```python
query = ''' 
SELECT * 
FROM fb55.citibike
ORDER BY cartodb_id
LIMIT 10 '''
get_data(query)
```

```python
query = ''' 
SELECT * 
FROM fb55.citibike
ORDER BY cartodb_id DESC
LIMIT 10 '''
get_data(query)
```

#### List all unique start_station_id values 

```python
query = ''' 
SELECT DISTINCT start_station_id 
FROM fb55.citibike
LIMIT 5
'''
get_data(query)
```

#### Aggregation: 
- Count the number of trips and 
- find the average/min/max trip duration

```python
query = ''' 
SELECT COUNT(cartodb_id),
AVG(tripduration), 
MAX(tripduration), 
MIN(tripduration)
FROM fb55.citibike
'''
get_data(query)
```

## Task 2 — Working with date/time

#### Selecting trips started on Feb-02-2015 only 

```python
query = ''' 
SELECT * 
FROM fb55.citibike 
WHERE starttime >= '2015-02-02 00:00' AND starttime < '2015-02-02 23:59'
LIMIT 5
'''
get_data(query)
```

#### Selecting trips started on the weekends
- What are average trip duration during weekends? 

```python
query = ''' 
SELECT * 
FROM fb55.citibike
WHERE EXTRACT(DOW FROM starttime) in (0,6)
LIMIT 5;
'''
get_data(query)
```

```python
query = ''' 
SELECT AVG(tripduration) 
FROM fb55.citibike
WHERE EXTRACT(DOW FROM starttime) IN (0,6)
LIMIT 5;
'''
get_data(query)
```

#### Can we do the same for weekday?

```python
query = ''' 
SELECT AVG(tripduration) 
FROM fb55.citibike
WHERE EXTRACT(DOW FROM starttime) IN (1,2,3,4,5)
LIMIT 5;
'''
get_data(query)
```

## Task 3 — Working with Space
####  Showing the list of start station locations
- Using GROUP BY 

```python
query = ''' 
SELECT start_station_name 
FROM fb55.citibike
GROUP BY start_station_name
LIMIT 5 
'''
get_data(query)
```

#### OR

```python
query = ''' 
SELECT start_station_latitude, start_station_longitude 
FROM fb55.citibike
GROUP BY start_station_longitude, start_station_latitude
LIMIT 5
'''
get_data(query)
```

#### Showing the number of trips started per station 

```python
query = ''' 
SELECT start_station_name, COUNT(start_station_name) 
FROM fb55.citibike
GROUP BY start_station_name
LIMIT 5 
'''
get_data(query)
```

### … but only for stations within 500m of Time Square!
- The coordinates of Time Square is (40.7577,-73.9857)

```python
query = '''
SELECT start_station_name, count(start_station_name)
FROM fb55.citibike
WHERE (6371000 * acos(sin(radians(40.7577)) * sin(radians(start_station_latitude)) + 
cos(radians(40.7577)) * cos(radians(start_station_latitude)) * cos(radians(start_station_longitude) - radians(-73.9857)))) <= 500
GROUP BY start_station_name
'''
  
get_data(query)
```

## Task 4 — Putting it all together
#### Find the station that had the longest average trip duration during weekends and within 500m of TimeSquare!

```python
query = '''
SELECT start_station_name, AVG(tripduration) as avg_duration
FROM fb55.citibike

WHERE (6371000 * acos(sin(radians(40.7577)) * sin(radians(start_station_latitude)) + 
cos(radians(40.7577)) * cos(radians(start_station_latitude)) * 
cos(radians(start_station_longitude) - radians(-73.9857)))) <= 500 
AND EXTRACT (dow FROM starttime) IN (0,6)

GROUP BY start_station_name
ORDER BY avg_duration DESC
LIMIT 1
'''
  
get_data(query)
```

#### Extra: create lines for trips started from stations within 500m of Times Squares and lasted less than 2 hours. The number of trips per each pair of stations are output as attributes of these lines. 

```python
query = '''
SELECT start_station_name, end_station_name, COUNT(tripduration)
FROM fb55.citibike

WHERE (6371000 * acos(sin(radians(40.7577)) * sin(radians(start_station_latitude)) + 
cos(radians(40.7577)) * cos(radians(start_station_latitude)) * 
cos(radians(start_station_longitude) - radians(-73.9857)))) <= 500 
AND tripduration < 7200 

GROUP BY start_station_name, end_station_name
LIMIT 5
'''
  
get_data(query)
```
